### 1. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{(y_{i} - (b + mx_{i}))^{2}}$$

#### &emsp;&emsp;де x,y - деякі задані вектори

```python
np.random.seed(42)
x = np.random.normal(0, 5, 100)
y = 50 + 2 * x + np.random.normal(0, 2, len(x))
```

```python
def f(b, m, x, y):
    g = (y-(b+m*x))**2
    return sum(g)/len(g)
```

```python
def df(b, m, x, y):
    h=1e-10
    f_b= (f(b+h, m, x, y)-f(b, m, x,y)) / h
    f_m= (f(b, m+h, x, y)-f(b, m, x,y)) / h
    return np.array([f_b, f_m])
```

```python
w0 = np.zeros(2)
alpha = 0.01
epsilon = 0.00001

print(w0, f(w0[0], w0[1], x, y))

while True:
    b = w0[0]
    m = w0[1]    
    w = w0 - alpha * df(b, m, x, y)
    
    if np.sum(np.abs(w-w0)) <= epsilon:
        break
    else:
        w0 = w
    
    print(w, f(w[0], w[1], x, y))
    
print()
print('w from f is min in : ', w)
print('Min f: ', f(w[0], w[1], x, y))
```

### 2. Знайти значення m і b для яких функція набуває мінімального значення:

$$f = \frac{1}{n}\sum_{i=1}^{n}{[(y_{i} - (b + mx_{i}))^{2}]} + m^2 + b^2$$

#### &emsp;&emsp;де x,y - вектори з 1-го завдання

```python
def f(b, m, x, y):
    g = (y-(b+m*x))**2
    return (sum(g)/len(g))+m**2+b**2
```

```python
def df(b, m, x, y):
    h=1e-10
    f_b= (f(b+h, m, x, y)-f(b, m, x,y)) / h
    f_m= (f(b, m+h, x, y)-f(b, m, x,y)) / h
    return np.array([f_b, f_m])
```

```python
w0 = np.zeros(2)
alpha = 0.0001
epsilon = 0.00001

print(w0, f(w0[0], w0[1], x, y))

while True:
    b = w0[0]
    m = w0[1]    
    w = w0 - alpha * df(b, m, x, y)
    
    if np.sum(np.abs(w-w0)) <= epsilon:
        break
    else:
        w0 = w
    
    print(w, f(w[0], w[1], x, y))
    
print()
print('w from f is min in : ', w)
print('Min f: ', f(w[0], w[1], x, y))
```

### 3. Знайти значення а, для якого функція набуває мінімального значення

$$f = -\frac{1}{n}\sum_{i=1}^{n}{[y_{i}log(h(x_{i})) + (1 - y_{i})log(1 - h(x_{i}))]}$$

$$h = \frac{1} {1 + exp^{-ax}}$$

```python
np.random.seed(10)
num_observations = 100

x1 = np.random.normal(1, 1, 50)
x2 = np.random.normal(3, 1, 50)

x = np.concatenate([x1, x2])
y = np.concatenate([np.zeros_like(x1), np.ones_like(x2)])
```

```python
def h(x, a):
    var = -a*x
    return 1/(1+np.exp(var))
```

```python
def f(x, a, y):
    g = y*np.log(h(x, a))+(1-y)*np.log(1-h(x, a))
    return -1*(sum(g)/len(g))
```

```python
def df(x, a, y):
    r = 1e-10
    return (f(x, a+r, y) - f(x, a, y)) / r 
```

```python
w0 = np.zeros(1)
alpha = 0.0001
epsilon = 0.00001

print(w0, f(x, w0, y))

while True:
    
    w = w0 - alpha * df(x, w0, y)
    
    if np.sum(np.abs(w - w0)) <= epsilon:
        break
    else:
        w0 = w
            
    print(w, f(x, w, y))
    
print()
print('w from f is min in : ', w)
print('Min f: ', f(x, w0, y))
```
