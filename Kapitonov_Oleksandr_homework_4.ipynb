### 1. Вибрати імена пасажирів, які підпадають під такі критерії:
#### - Клас каюти 1-й чи 3-й,
#### - Вік в інтервалі від 20 до 40,
#### - Ціна квитка НЕ знаходиться в інтервалі від 70 до 100.

```python
# Клас каюти 1-й чи 3-й
cabin =  titanic_data[(titanic_data['Pclass']==1) | (titanic_data['Pclass']==3)]['Name']
cabin
```

```python
# Вік в інтервалі від 20 до 40
age_inter = titanic_data[(titanic_data['Age']<40) & (titanic_data['Age']>20)]['Name']
age_inter
```

```python
# Ціна квитка НЕ знаходиться в інтервалі від 70 до 100
price_inter = titanic_data[(titanic_data['Fare']<70) | (titanic_data['Fare']>100)]['Name']
price_inter
```

```python
# Ті хто є в усіх 3 категоріях 
all_3 = pd.merge(pd.merge(age_inter, cabin), price_inter)
all_3
```
